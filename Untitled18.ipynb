```python
import torch
from diffusers import StableDiffusionXLPipeline, AutoencoderKL
from diffusers.utils import load_image
import os
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    variant="fp16",
).to(device)
```

```python
suv_image = load_image("/content/car_angular.png")
```

```python
prompts = {
    "mumbai_sealink_wide": "A silver Mahindra XUV700 SUV on an empty road with the Bandra-Worli Sea Link in the background, wide cinematic view, natural daylight, ultra-realistic photograph, 8k sharpness",
    "mumbai_sealink_front": "A silver Mahindra XUV700 SUV facing the Bandra-Worli Sea Link entrance, front angle, cinematic HDR photography, realistic background",
    "hawa_mahal_wide": "A silver Mahindra XUV700 SUV parked in front of Jaipur’s Hawa Mahal, wide shot with full façade, sunset golden lighting, realistic architecture details",
    "hawa_mahal_side": "A silver Mahindra XUV700 SUV side profile parked at the street in front of Hawa Mahal, low dramatic angle, warm tones, realistic photo",
    "nagpur_rainforest_trail": "A silver Mahindra XUV700 SUV driving on a muddy rainforest trail near Nagpur, lush greenery, misty atmosphere, cinematic wide shot",
    "nagpur_rainforest_aerial": "A silver Mahindra XUV700 SUV captured from aerial angle in Nagpur rainforest, surrounded by dense jungle, wet ground, cinematic realism"
}
neg_prompt ={"unrealsitic","cartoon like"}
```

```python
os.makedirs("generated_suv_images", exist_ok=True)
```

```python
for name, prompt in prompts.items():
    image = pipe(
        prompt=prompt,
        negative_prompt=neg_prompt
        image=suv_image,
        num_inference_steps=30,
        guidance_scale=7.5,
    ).images[0]

    image.save(f"generated_suv_images/{name}.png")
    print(f"Saved {name}.png")
```
